**RETURN STATEMENT**

- return statements are used to return the values

- every function will do some operations

- those operations will save in 1 variable

- if u want to use those values out side the function  we can not use

- for that we are using **global** keyword

- we can also use those variable values out side function using **return**

```python
def avg5(a,b,c=50):

    summ=(a+b+c)
    avg=summ/3
    print(summ)
    return (avg,summ)
avg=avg5(10,20,10)
```
